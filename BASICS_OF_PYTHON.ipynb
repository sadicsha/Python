**FACTORIAL OF A NUMBER**

```
num = int(input("Enter a number: "))    
print(num)
factorial=1
for i in range(1,num+1):    
       factorial = factorial*i    
       print("The factorial of",num,"is",factorial)    
```

**PYTHON PROGRAM FOR SIMPLE INTREST AND COMPOUND INTREST**

```
time = int(input('Enter time: '))
rate = float(input('Enter rate: '))
principle = float(input('Enter principle: '))
Simple_intrest = (time*rate*principle)/100
print("Simple Intrest is: ", Simple_intrest)
Compound_intrest = principle*((1+rate)**time)-principle
print("Compound Intrest is: ",Compound_intrest)
```

**PRIME NUMBER**

```
num = int(input('Enter a number: '))
if num>1:
  for i in range(2,num):
    if (num%i)==0:
      print(num,"not prime")
      break
    else:
      print("is prime")
else:
  print("not prime")
    
```
